```julia
include("./src/localLinear_quantReg.ji")
include("./src/supportFunctions.ji")
using LinearAlgebra, Statistics
using Plots
```

```julia
n = 100
Xn = randn(n, 2)
X = [ones(n)  Xn]

b = [-1.0, 2.0, 0.5]

y = X*b + randn(n) * 0.2

h = var(X[:,2]) * n^(-0.17)

w = ker(X[:,2] .- randn(n), h)
```

```julia
plot(ker)
```

```julia
t1 = ρ.(X*b) .* w
t2 = ρ.(X*b .* w)
plot(t2 - t1)
```

```julia
using  QuantReg, DataFrames,GLM

n = 100
Xn = randn(n, 2)
X = [ones(n)  Xn]

b = [-1.0, 2.0, 0.5]

y = X*b + randn(n) * 0.4
df = DataFrame(Xn)
nox = Symbol.(names(df))
df[:y] = y
formula1 = Term(:y) ~ sum([Term(x) for x in nox])

model = QuantRegModel(formula1, df, τ = 0.5)
```

```julia
dof(model)
```
